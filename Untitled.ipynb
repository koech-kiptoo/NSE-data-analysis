```python
import matplotlib.pyplot as plt
import numpy as np

salesa=[2000,1200,5000,6000,3200,8000]
expenze=[1000,1000,4500,4000,2500,7500]
monthsb=["Jan","Feb","March","April","May","June"]

x_axis=np.arange(len(monthsb))

plt.bar(x_axis-0.2,salesa,width=0.3,label= 'salesa')
plt.bar(x_axis+0.2,expenze,width=0.4,label= 'Expenze')

plt.xticks(x_axis, monthsb)
plt.legend()
plt.show
```

```python
towns=["Makueni","Nairobi","Kiambu","Machakos"]
numberofhiv=[454,75,578,654]
numberofmalaria=[254,400,400,100]

plt.scatter(towns, numberofhiv)
plt.scatter(towns,numberofmalaria)
plt.show()
```

```python
towns= ["Kisumu", "Nairobi", "Eldoret", "Mombasa"]
numberofhiv = [454, 75, 578, 654]
numberofmalaria = [254, 400, 400, 100]

plt.scatter(towns, numberofhiv, label="HIV")
plt.scatter(towns, numberofmalaria, label="Malaria")

handles = {
    "HIV": plt.scatter([], [], c="blue"),
    "Malaria": plt.scatter([], [], c="orange")
}

plt.legend(handles=handles.values(), labels=handles.keys(),
           title="Disease", scatterpoints=1, fontsize=8,
           loc="upper left", bbox_to_anchor=(1.05, 1))
```
